```python
import json
import mysql.connector

# Load JSON data from file
with open('station_list.json') as f:
    data = json.load(f)

# Connect to MySQL
connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="duke19990",
    database="station"
)
cursor = connection.cursor()

# Create table based on JSON structure
table_creation_query = """
    CREATE TABLE IF NOT EXISTS station_list (
        id INT AUTO_INCREMENT PRIMARY KEY,
        station VARCHAR(50),
        station_code VARCHAR(50),
        state VARCHAR(50),
        railway_zone VARCHAR(50)
    )
"""
cursor.execute(table_creation_query)

# Insert data into the table
for item in data:
    insert_query = """
        INSERT INTO station_list (station, station_code,state,railway_zone)
        VALUES (%s, %s, %s, %s)
    """
    cursor.execute(insert_query, (item['station'], item['station_code'], item['state'], item['railway_zone']))

# Commit changes and close connection
connection.commit()
connection.close()
```

```python
import json
import mysql.connector

# Load JSON data from file
with open('trains.json') as f:
    data = json.load(f)
```

```python
import pandas as pd
import re

# Assuming df is your DataFrame and 'Train No' is the column name
# Replace 'Train No' with your actual column name if it's different

# Convert dictionaries to strings
df['Train No'] = df['Train No'].apply(lambda x: x.get(''))

# Apply regular expression to extract only numbers
df['Train No'] = df['Train No'].apply(lambda x: re.sub(r'\D', '', str(x)))
```
